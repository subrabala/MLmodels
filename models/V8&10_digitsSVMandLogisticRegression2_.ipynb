```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
plt.matshow(digits.images[1])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (digits.data, digits.target,train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
# plt.gray()
# plt.matshow(y.images[0])
```

```python
model.score(x_test, y_test)
```

```python
plt.gray()
plt.matshow(digits.images[45])
```

```python
model.predict(x_train[[45]])
```

```python
digits.target[45]
```

```python
# target = actual number
# image = binary data
# data = 8*8 array values
```

```python
y_predicted = model.predict(x_test)
from sklearn.metrics import confusion_matrix
conf = confusion_matrix(y_test,y_predicted)
conf
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(conf,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.svm import SVC
svc_model =SVC(kernel="linear")
```

```python
svc_model.fit(x_train,y_train)
```

```python
svc_model.score(x_test,y_test)
```

```python
y_pred_svc= svc_model.predict(x_test)
```

```python
conf_svc = confusion_matrix(y_test,y_pred_svc)
plt.figure(figsize = (10,7))
sn.heatmap(conf_svc,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
